```python
#read the ratings dataset
r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=r_cols,encoding='latin-1')
```

```python
# Create Pivot Table  which will act as user vector
datama=ratings.pivot_table(index='user_id',columns='movie_id',values='rating')
datama
```

```python
data_matrix=datama.replace(np.nan,0)
data_matrix
```

```python
user_vector=data_matrix.T
user_vector
```

```python
# Create movie Vector

movie_vec=items[['Action', 'Adventure',
    'Animation','Childre\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']]
```

```python
movie_vec
```

```python
# Rename the index
movie_vec.index=range(1,1683)
movie_vec
```

```python
movie_vector=movie_vec.T
movie_vector
```

```python
# Calculate average rating per movie
movie_avg_rating = ratings.groupby('movie_id')['rating'].mean().reset_index()

# Rename column for clarity
movie_avg_rating.rename(columns={'rating': 'avg_rating'}, inplace=True)

# Sort movies by average rating (descending)
movie_avg_rating = movie_avg_rating.sort_values(by='avg_rating', ascending=False)

print(movie_avg_rating.head(10))
```

```python
import pandas as pd

# Suppose movie_vector is numpy array with shape (num_movies, num_features)
movie_vector_df = pd.DataFrame(movie_vector.T)

# Add movie_id column (must match the IDs in ratings/movies metadata)
movie_vector_df['movie_id'] = range(1, len(movie_vector_df) + 1)
```

```python
movie_with_vector = pd.merge(movie_avg_rating, movie_vector_df, on='movie_id')
```

```python
movie_with_vector
```

```python
import pandas as pd

genres = [
    'Action', 'Adventure', 'Animation', 'Childre\'s', 'Comedy', 'Crime',
    'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical',
    'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western'
]
# Collect results in a list
results = []

for genre in genres:
    top5 = (
        movie_with_vector[movie_with_vector[genre] == 1]
        .sort_values(by='avg_rating', ascending=False)
        .head(5)
        .copy()
    )
    top5['genre'] = genre   # add genre column
    results.append(top5[['genre', 'movie_id', 'avg_rating']])

# Combine into one DataFrame
final_table = pd.concat(results, ignore_index=True)

print(final_table)
```
