# Create pivot table for user and movie basedon ratings

```python
datama=ratings.pivot_table(index='user_id',columns='movie_id',values='rating')
datama
```

```python
#replace with nan with 0
data_matrix=datama.replace(np.nan,0)
data_matrix
```

Find cosine similarity for user and item

```python
from sklearn.metrics.pairwise import pairwise_distances
user_similarity=pairwise_distances(data_matrix,metric='cosine')
item_similarity=pairwise_distances(data_matrix.T,metric='cosine')
```

```python
print(user_similarity)
```

# Using formula for user and item we are calcuating the score value

```python
def predict(ratings,similarity,type='user'):
    ratings=np.array(ratings)
    similarity=np.array(similarity)
    if type=='user':
        mean_user_rating=ratings.mean(axis=1)#-->Calculating avg. row wise.
        mean_user_rating=np.array(mean_user_rating)
        rating_diff=ratings-mean_user_rating[:,np.newaxis]#--> converting into matrix as it was in the original data by using np.newaxis.
        pred=mean_user_rating[:,np.newaxis]+similarity.dot(rating_diff)/np.array([np.abs(similarity).sum(axis=1)]).T#-->Convert 1d array into column vector, rating diff(diff/ between rating and mean,similarity matrix multiplication/
    elif type=='item':
        pred=ratings.dot(similarity)/np.array([np.abs(similarity).sum(axis=1)])
    return pred
```

```python
user_prediction=predict(data_matrix,user_similarity,type='user')
item_prediction=predict(data_matrix,item_similarity,type='item')
```
